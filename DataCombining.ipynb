```python
import pandas as pd
import numpy as np
import ast

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.preprocessing import MultiLabelBinarizer
```

# Initial data processing

```python
Kaggle_data = pd.read_csv("Kaggle_data.csv", encoding="latin-1")
```

```python
Mined_data = pd.read_csv("Mined_data.csv", encoding="latin-1")
```

```python
num_cols = [
    "All Time Rank",
    "Track Score",
    "Spotify Streams",
    "Spotify Playlist Count",
    "Spotify Playlist Reach",
    "Spotify Popularity",
    "YouTube Views",
    "YouTube Likes",
    "TikTok Posts",
    "TikTok Likes",
    "TikTok Views",
    "YouTube Playlist Reach",
    "Apple Music Playlist Count",
    "AirPlay Spins",
    "SiriusXM Spins",
    "Deezer Playlist Count",
    "Deezer Playlist Reach",
    "Amazon Playlist Count",
    "Pandora Streams",
    "Pandora Track Stations",
    "Soundcloud Streams",
    "Shazam Counts"
]

streams = ["Spotify Streams", "AirPlay Spins", "SiriusXM Spins", "Pandora Streams", "Soundcloud Streams"]
views = ["YouTube Views", "TikTok Views"]
likes = ["YouTube Likes", "TikTok Likes"]
playlist_counts = ["Spotify Playlist Count", "Apple Music Playlist Count", "Deezer Playlist Count", "Amazon Playlist Count", "Pandora Track Stations"]
playlist_reach = ["Spotify Playlist Reach", "YouTube Playlist Reach", "Deezer Playlist Reach"]
```

```python
Kaggle_data = Kaggle_data.drop("TIDAL Popularity", axis=1)
Kaggle_data[num_cols] = Kaggle_data[num_cols].replace(0, np.nan)

for col in num_cols:
    Kaggle_data[col] = Kaggle_data[col].astype(str).str.replace(",", "").astype(float)

Kaggle_data["Release Date"] = pd.to_datetime(Kaggle_data["Release Date"])
Kaggle_data["Explicit Track"] = Kaggle_data["Explicit Track"].astype(bool)
```

```python
Mined_data = Mined_data.drop("New Name", axis=1)
Mined_data = Mined_data.drop("Id", axis=1)
Mined_data = Mined_data.dropna(subset=["Artist"])
```

# Datasets combining

```python
Combined_data = pd.merge(Kaggle_data, Mined_data, on="Artist", how="left")
```

```python
Combined_data["Artist Genres"] = Combined_data["Artist Genres"].fillna("[]")
Combined_data["Artist Genres"] = Combined_data["Artist Genres"].apply(ast.literal_eval)
```

```python
Combined_data.info()
```

# Attributes combining by normalizing

```python
Normalized_data = Combined_data.copy()
scaler = MinMaxScaler()
Normalized_data[num_cols] = scaler.fit_transform(Normalized_data[num_cols])
```

```python
Normalized_data["Streams"] = Normalized_data[streams].mean(axis=1)
Normalized_data["Views"] = Normalized_data[views].mean(axis=1)
Normalized_data["Likes"] = Normalized_data[likes].mean(axis=1)
Normalized_data["Playlist Count"] = Normalized_data[playlist_counts].mean(axis=1)
Normalized_data["Playlist Reach"] = Normalized_data[playlist_reach].mean(axis=1)
```

```python
Normalized_data.to_csv("Normalized_data.csv", index=False)
```

# Attributes combining by standardizing

```python
Standardized_data = Combined_data.copy()
scaler = StandardScaler()
Standardized_data[num_cols] = scaler.fit_transform(Standardized_data[num_cols])
```

```python
Standardized_data["Streams"] = Standardized_data[streams].mean(axis=1)
Standardized_data["Views"] = Standardized_data[views].mean(axis=1)
Standardized_data["Likes"] = Standardized_data[likes].mean(axis=1)
Standardized_data["Playlist Count"] = Standardized_data[playlist_counts].mean(axis=1)
Standardized_data["Playlist Reach"] = Standardized_data[playlist_reach].mean(axis=1)
```

```python
Standardized_data.to_csv("Standardized_data.csv", index=False)
```

# Attributes combining for analysis

```python
Averaged_data = Combined_data.copy()

Averaged_data["Streams"] = Averaged_data[streams].mean(axis=1)
Averaged_data["Views"] = Averaged_data[views].mean(axis=1)
Averaged_data["Likes"] = Averaged_data[likes].mean(axis=1)
Averaged_data["Playlist Count"] = Averaged_data[playlist_counts].mean(axis=1)
Averaged_data["Playlist Reach"] = Averaged_data[playlist_reach].mean(axis=1)
```
