```python
!pip install spotipy
import spotipy
import json
import pandas as pd

from spotipy.oauth2 import SpotifyClientCredentials
```

```python
client_id = "ENTER YOUR ID"
client_secret = "ENTER YOUR SECRET"

auth_manager = SpotifyClientCredentials(
    client_id=client_id,
    client_secret=client_secret
)

sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
df = pd.read_csv("Kaggle_data.csv", encoding="latin-1")
```

```python
def get_artist_id(name):
    result = sp.search(q=name, type="artist", limit=1)
    items = result["artists"]["items"]
    if items:
        return items[0]["id"]
    else:
        return None

def get_artist_genre(id):
    if id is None:
        return None, None
    result = sp.artist(id)
    return result["genres"], result["name"]
```

```python
datamine = pd.DataFrame(columns = ["Artist", "Id", "Artist Genres", "New Name"])

datamine["Artist"] = df["Artist"].unique()
datamine["Id"] = datamine["Artist"].apply(get_artist_id)
datamine[["Artist Genres", "New Name"]] = datamine["Id"].apply(lambda x: pd.Series(get_artist_genre(x)))
```
