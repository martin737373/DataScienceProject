```python
df = pd.read_csv("Kaggle_data.csv", encoding="latin-1")
df
```

```python
def get_artist_id(name):
    result = sp.search(q=name, type="artist", limit=1)
    items = result["artists"]["items"]
    if items:
        return items[0]["id"]
    else:
        return None

def get_artist_genre(id):
    if id is None:
        return None, None
    result = sp.artist(id)
    return result["genres"], result["name"]
```

```python
datamine = pd.DataFrame(columns = ["name", "id", "genres", "new_name"])
datamine["name"] = df["Artist"].unique()
datamine["id"] = datamine["name"].apply(get_artist_id)
datamine[["genres", "new_name"]] = datamine["id"].apply(lambda x: pd.Series(get_artist_genre(x)))
```
